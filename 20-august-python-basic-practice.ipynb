```python
# 42. Encode a string using a Caesar cipher.
def encode_caesar(string, shift):
    new = list(string.lower())
    en_list = []
    for i in range(len(new)):
        if new[i].isalpha():
            message = chr((ord(new[i])+ shift - ord('a')) % 26 + ord('a'))
            en_list.append(message)
        else:
            en_list.append(new[i])
    return ''.join(en_list)

string = "I will takeover the pw"
shift = 5
print(f"The encoded string is: {encode_caesar(string, shift)}")
```

```python
# 43. Decode a Caesar cipher encoded string.
def decode_caesar(string, shift):
    
    new = list(string)
    en_list = []
    for i in range(len(new)):
        if new[i].isalpha():
            message = chr((ord(new[i])- shift - ord('a')) % 26 + ord('a'))
            en_list.append(message)
        else:
            en_list.append(new[i])
    return (''.join(en_list)).title()

# encoded message: n bnqq yfpjtajw ymj ub
string = input("Enter an Encoded message: ")
shift = 5
print(f"The decoded string is: {decode_caesar(string, shift)}")
```

```python
# 31. Rotate a list to the left by `n` positions.
def rotate_left(list, rotate):
    rotate = rotate % len(list)
    return list[rotate:] + list[:rotate]

list = [11, 21, 31, 41, 51]
rotate = 3
print(f"list after left rotation: {rotate_left(list, rotate)}")
```

```python
# 32. Rotate a list to the right by `n` positions.
def rotate_right(list, rotate):
    rotate = rotate % len(list)
    return list[-rotate:] + list[:-rotate]

list = [11, 21, 31, 41, 51]
rotate = 3
print(f"list after right rotation: {rotate_right(list, rotate)}")
```

```python
# 33. Create a list of prime numbers up to 50.
def primes(n):
    primes = []
    for i in range(2, n + 1):
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            primes.append(i)
    return primes

n = 50
print(f"list of prime numbers up to {n}: {primes(n)}")
```

```python
# 41. Check if the intersection of two sets is empty.
def check_intersection_empty(set1, set2):
    return set1.intersection(set2) == set()

set36 = {1, 2, 3, 4, 5}
set37 = {6, 7, 8, 9, 10}
print(f"intersection of two sets is empty: {check_intersection_empty(set36, set37)}")
```

```python
# 42. Create a set of the first 5 Fibonacci numbers.
def create_fibonacci_set(n):
    a, b = 0, 1
    fibonacci_set = set()
    for _ in range(n):
        fibonacci_set.add(a)
        a, b = b, a + b
  return fibonacci_set

n = 5
print(f"set of the first 5 Fibonacci numbers: {create_fibonacci_set(n)}")
```

```python
# 44. Check if two sets have the same elements, regardless of their count.
def check_same_elements(set1, set2):
    return set1.issubset(set2) and set2.issubset(set1)

set38 = {1, 2, 3, 4, 5}
set39 = {3, 4, 5, 1, 2, 4, 2, 5}
print(f"two sets have the same elements, regardless of their count: {check_same_elements(set38, set39)}")
```

```python
# 45. Create a set of the first `n` powers of 2.
def create_power_set(n):
    return {2 ** i for i in range(n+1)}

n = 5
print(f"set of the first {n} powers of 2: {create_power_set(n)}")
```

```python
# 46. Find the common elements between a set and a list.
def find_common_elements(set40, list40): 
    convert_list = set(list40)
    common = []
    for i in set40:
        if i in convert_list:
            common.append(i)
    return common

set40 = {1, 2, 3, 4, 5}
list40 = [3, 4, 5, 6, 7]
print(f"common elements between a set and a list: {find_common_elements(set40, list40)}")
```

```python
# 47. Create a set of the first `n` triangular numbers.
def create_triangular_set(n):
    return {i * (i + 1) // 2 for i in range(1, n+1)}

n = 5
print(f"set of the first {n} triangular numbers: {create_triangular_set(n)}")
```

```python
# 48. Check if a set contains another set as a subset.
def check_subset(set2, set1):
    return set1.issubset(set2)

set41 = {1, 2, 3, 4, 5}
set42 = {3, 4, 5}
print(f"set41 contains set42 as a subset: {check_subset(set41, set42)}")
```

```python
# 49. Create a set of alternating 1s and 0s of length `n`.
def create_alternating_set(n):
    return {0 if i%2 == 0 else 1 for i in range(n)}

n = 5
print(f"set of alternating 1s and 0s of length {n}: {create_alternating_set(n)}")
```

```python
# 50. Merge multiple sets into one.
def merge_sets(*sets):
    return set.union(*sets)

set43 = {1, 2, 3}
set44 = {4, 5, 6}
set45 = {7, 8, 9}
print(f"merge multiple sets into one: {merge_sets(set43, set44, set45)}")
```
